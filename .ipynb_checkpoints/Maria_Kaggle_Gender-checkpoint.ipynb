```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
gender_data = pd.read_csv('Resources/kaggle_gender.csv')
gender_data
```

```python
gender_data=gender_data[gender_data.M_weekly!='Na']
gender_data=gender_data[gender_data.F_weekly!='Na']
gender_data=gender_data[gender_data.All_workers!='Na']
```

```python
gender_data.dtypes
```

```python
#pull columns out into arrays to convert
occupation=gender_data.iloc[1:,0].values
m_salary=gender_data.iloc[1:,4].values
f_salary=gender_data.iloc[1:,6].values
all_workers=gender_data.iloc[1:,1].values
m_workers=gender_data.iloc[1:,3].values
f_workers= gender_data.iloc[1:,5].values
all_salary=gender_data.iloc[1:,2].values
```

```python
m_salary=pd.to_numeric(m_salary,downcast='integer')
f_salary=pd.to_numeric(f_salary,downcast='integer')
all_workers=pd.to_numeric(all_workers,downcast='integer')
m_workers=pd.to_numeric(m_workers,downcast='integer')
f_workers=pd.to_numeric(f_workers,downcast='integer')
all_salary=pd.to_numeric(all_salary,downcast='integer')
```

```python
#encode occupation label
from sklearn.preprocessing import LabelEncoder
label_x=LabelEncoder()
occupation_number=label_x.fit_transform(occupation)
```

```python
#add columns to df
new_gender=pd.DataFrame()
new_gender['occupation_number']=occupation_number
new_gender['occupation']=occupation
new_gender['all_workers']=all_workers
new_gender['all_salary']=all_salary
new_gender['male_workers']=m_workers
new_gender['male_salary']=m_salary
new_gender['female_workers']=f_workers
new_gender['female_salary']=f_salary
new_gender['female_to_male_salary_ratio'] = new_gender.female_salary/new_gender.male_salary
new_gender['female_to_male_workers_ratio'] = new_gender.female_workers/new_gender.male_workers
new_gender['male_worker_percent']= new_gender.male_workers/new_gender.all_workers
new_gender['female_worker_percent']= new_gender.female_workers/new_gender.all_workers
new_gender['representation_gap']= new_gender.male_workers - new_gender.female_workers
```

```python
#when worker ratio is greater than 1, male probability is higher

prob=[]
array=new_gender['female_to_male_workers_ratio'].values
for i in array:
    if i>1:
        prob.append(0)
    else:
        prob.append(1)
new_gender['male_probability']=prob
```

```python
new_gender = new_gender.drop(columns=['male_female_prob'])
```

```python
#sort 

sorted_gender = new_gender.sort_values(['female_to_male_workers_ratio'], ascending = True)
```

```python
#KNN with male probability
from sklearn.neighbors import KNeighborsClassifier
```

```python
new_gender = new_gender.drop(columns=["occupation"])
```

```python
y = new_gender["male_probability"]
target_names = ["improbable", "probable"]
```

```python
X = new_gender.drop("male_probability",axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

X_scaler = StandardScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler and y_scaler models

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.title("KNN Model for Predicting Male Probability by Profession")
plt.show()
```

```python
# Note that k: 3 seems to be the best choice for this dataset
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
print('k=3 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```
