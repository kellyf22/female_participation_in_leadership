```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sf_data = pd.read_csv('Resources/States_Data/merged_states.csv')
sf_data.head()
```

```python
narrow_df = sf_data
```

```python
narrow_df
```

```python
management_df = narrow_df[narrow_df['Label'].str.contains("Management")]
management_df
```

```python
management_df = management_df.drop(columns=["Label", "US_Total"])
```

```python
management_df
```

```python
#add column for gap
management_df['Percent_Salary_Gap']= management_df.US_Male_Percent - management_df.US_Female_Percent
```

```python
management_df
```

```python
management_df = management_df.drop(columns=["percent_salary_gap"])
```

```python
#drop US MALE and FEMALE
salary_gap_df = management_df.drop(columns=["US_Male_Percent", "US_Female_Percent"])
```

```python
salary_gap_df = salary_gap_df.reset_index(drop=True)
```
