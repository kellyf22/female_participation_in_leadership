```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sf_data = pd.read_csv("Resources/States_Data/merged_states.csv")
sf_data
```

```python
narrow_df = sf_data.drop(columns=['US_Total_Earnings_Male','US_Total_Earnings_Female','US_Total','US_Median_Earnings_Male','US_Median_Earnings_Female','US_Total_Median_Earnings '])
narrow_df
```

```python
narrow_df = narrow_df[narrow_df['Label'].str.contains("Management")]
narrow_df
```

```python
narrow_df['US_Male_Percent'] = narrow_df['US_Male_Percent'].str.rstrip('%').astype('float') / 100.0
narrow_df
```

```python
narrow_df['US_Female_Percent'] = narrow_df['US_Female_Percent'].str.rstrip('%').astype('float') / 100.0 
narrow_df
```

```python
#Can we predict year of income equality using percent of male and female income? 
X = narrow_df[["US_Male_Percent", "US_Female_Percent"]]
y = narrow_df["Year"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data
### BEGIN SOLUTION
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data
### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
```

```python
print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data

### BEGIN SOLUTION
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
### END SOLUTION
```

```python
### BEGIN SOLUTION
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
### END SOLUTION
```

```python
### BEGIN SOLUTION
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
### END SOLUTION
```

```python
predictions = model.predict(X_test_scaled)
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), y_train_scaled - model.predict(X_train_scaled), c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), y_test_scaled - model.predict(X_test_scaled), c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()
```
