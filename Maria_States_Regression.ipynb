```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python

sf_data = pd.read_csv('Resources/States_Data/merged_states.csv')
sf_data.head()
```

```python
narrow_df = sf_data.drop(columns=['US_Total_Median_Earnings ','US_Median_Earnings_Male', 'US_Median_Earnings_Female', 'US_Total_Earnings_Male','US_Total_Earnings_Female'])
narrow_df
```

```python
narrow_df['US_Male_Percent'] = narrow_df['US_Male_Percent'].str.rstrip('%').astype('float') / 100.0
```

```python
narrow_df['US_Female_Percent'] = narrow_df['US_Female_Percent'].str.rstrip('%').astype('float') / 100.0
```

```python
narrow_df.dtypes
```

```python
#filter out management positions
management_df = narrow_df[narrow_df['Label'].str.contains("Management")]
```

```python
management_df
```

```python
#drop label for the model

management_df = management_df.drop(columns=["Label", "US_Total"])
```

```python
management_df
```

```python
sns.relplot(x='Year', y='US_Female_Percent', data=management_df)
```

```python
sns.relplot(y='US_Male_Percent', x='Year', data=management_df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Assign the data to X and y

X = management_df[["US_Male_Percent", "US_Female_Percent"]]
y = management_df["Year"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create the model using LinearRegression

### BEGIN SOLUTION
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Plot the Residuals for the Training and Testing data

### BEGIN SOLUTION
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
### END SOLUTION
```
